```python
import requests 
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36"
}
```

```python
url = 'https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
response = requests.get(url,headers = headers)
```

```python
response.status_code
```

```python
html_data = response.text
soup = BeautifulSoup(html_data)
```

```python
name = soup.find_all("div",class_ = "RG5Slk")
```

```python
product_name = []
for i in name:
    product_name.append(i.text)
```

```python
product_name 
```

```python
Brand = []
Processor = []
RAM = []
SSD = []
Display = []
Price = []
Discount = []
Rating = []
data = soup.find_all("li",class_ = "DTBslk")
for i in data:
    if "Processor" in i.text:
        Processor.append(i.text)
    elif "RAM" in i.text:
        RAM.append(i.text)
    elif "SSD" in i.text:
        SSD.append(i.text)
    elif "Display" in i.text:
        Display.append(i.text)
        
price = soup.find_all("div",class_="hZ3P6w DeU9vF")
for i in price :
    Price.append(i.text)

rating = soup.find_all("div",class_="MKiFS6")
for i in rating:
    Rating.append(i.text)
discount = soup.find_all('div',class_ = "HQe8jr")
for i in discount:
    Discount.append(i.text)
for i in product_name:
    i = i.split(" ")
    Brand.append(i[0])
```

```python
print(len(product_name))
print(len(Brand))
print(len(Processor))
print(len(RAM))
print(len(SSD))
print(len(Display))
print(len(Price))
print(len(Discount))
print(len(Rating))
```

```python
min_len=min(len(product_name),
len(Brand),
len(Processor),
len(RAM),
len(SSD),
len(Display),
len(Price),
len(Discount),
len(Rating))
```
