```python
import requests 
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36"
}
```

```python
url = 'https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
response = requests.get(url,headers = headers)
```

```python
response.status_code
```

```python
html_data = response.text
soup = BeautifulSoup(html_data)
```

```python
name = soup.find_all("div",class_ = "RG5Slk")
```

```python
product_name = []
for i in name:
    product_name.append(i.text)
```

```python
product_name 
```

```python
Brand = []
Processor = []
RAM = []
SSD = []
Display = []
Price = []
Discount = []
Rating = []
data = soup.find_all("li",class_ = "DTBslk")
for i in data:
    if "Processor" in i.text:
        Processor.append(i.text)
    elif "RAM" in i.text:
        RAM.append(i.text)
    elif "SSD" in i.text:
        SSD.append(i.text)
    elif "Display" in i.text:
        Display.append(i.text)
        
price = soup.find_all("div",class_="hZ3P6w DeU9vF")
for i in price :
    Price.append(i.text)

rating = soup.find_all("div",class_="MKiFS6")
for i in rating:
    Rating.append(i.text)
discount = soup.find_all('div',class_ = "HQe8jr")
for i in discount:
    Discount.append(i.text)
for i in product_name:
    i = i.split(" ")
    Brand.append(i[0])
```

```python
df=pd.DataFrame({
    "product_name":product_name[:min_len],
    "Brand":Brand[:min_len],
    "Processor":Processor[:min_len],
    "RAM":RAM[:min_len],
    "SSD":SSD[:min_len],
    "Display":Display[:min_len],
    "Price":Price[:min_len],
    "Discount":Discount[:min_len],
    "Rating":Rating[:min_len] 
})
```

```python
df['RAM'] = df['RAM'].apply(lambda x: re.findall(r'(\d+)\s(?:GB|MB|LPDDR)', str(x))[0] if re.findall(r'(\d+)\s[GB]', str(x)) else x)
df['RAM'] = df['RAM'].apply(lambda x: f"{x} GB" if str(x).isdigit() else x)
```

```python
df["Display"] = (
    df["Display"]
    .str.extract(r"\(([\d.]+)\s*[Ii]nch\)", expand=False)
    + " inch"
)
```

```python
df["Price"] = df["Price"].apply(
    lambda x: re.findall(r"\d+[,]\d+", x)[0]
)
```

```python
df["Price"] = df["Price"].str.replace(",", "", regex=False)
```

```python
df['Price'] = df['Price'].astype(int)
```

```python
df['Rating'] = df['Rating'].astype(float)
```

```python
df.info()
```

```python
df.shape
```

```python
# checking the duplicates
df.duplicated().sum()
```

```python
# removing the duplicates
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
def category(price):
    if price < 25000:
        return "Budget"
    elif price < 50000:
        return "Mid Range"
    else:
        return "Premium"

df["Price_Category"] = df["Price"].apply(category)
```

```python
df["New_Discount"] = (
    df["Discount"]
    .str.replace("% off", "", regex=False)
    .astype(float)
)
```

```python
df["Original_Price"] = (
    df["Price"] * 100 / (100 - df["New_Discount"])
).round().astype(int)
```

```python
df['Processor_brand'] = df['Processor'].apply(
    lambda x: re.findall(r"\w+", x)[0]
)
```

## Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Price Distribution — Univariate Analysis

```python
plt.figure(figsize=(10, 5))

sns.histplot(
    data=df,
    x="Price",
    bins=30,
    kde=True
)

plt.title("Distribution of Laptop Prices")
plt.xlabel("Selling Price")
plt.ylabel("Number of Laptops")
plt.show()
```

### 2. Top 10 Brands — Univariate Analysis

```python
top_brands = df["Brand"].value_counts().head(10)

plt.figure(figsize=(10, 5))

sns.barplot(
    x=top_brands.values,
    y=top_brands.index
)

plt.title("Top 10 Brands by Number of Laptops")
plt.xlabel("Number of Laptops")
plt.ylabel("Brand")
plt.show()
```

### 3. RAM Distribution — Univariate Analysis

```python
plt.figure(figsize=(8, 5))

sns.countplot(
    data=df,
    x="RAM",
    order=df["RAM"].value_counts().index
)

plt.title("Distribution of RAM")
plt.xlabel("RAM (GB)")
plt.ylabel("Number of Laptops")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

sns.histplot(
    data=df,
    x="Rating",
    bins=30,
    kde=True
)

plt.title("Distribution of Laptop by Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Laptops")
plt.show()
```

### 4. Processor Comparison — Bivariate Analysis

```python
processor_price = (
    df.groupby("Processor")["Price"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(
    data=processor_price,
    x="Price",
    y="Processor"
)

plt.title("Top 10 Processors by Average Laptop Price")
plt.xlabel("Average Selling Price")
plt.ylabel("Processor")
plt.show()
```

### 5. RAM Comparison — Bivariate Analysis

```python
ram_price = (
    df.groupby("RAM")["Price"]
    .mean()
    .sort_values()
    .reset_index()
)

plt.figure(figsize=(10, 5))

sns.barplot(
    data=ram_price,
    x="RAM",
    y="Price"
)

plt.title("Average Laptop Price by RAM")
plt.xlabel("RAM (GB)")
plt.ylabel("Average Selling Price")
plt.show()
```

### 6. Price vs Rating — Bivariate Analysis

### 7. Correlation Heatmap — Multivariate Analysis

```python
columns = [
    "Original_Price",
    "Price",
    "New_Discount",
    "Rating"
]

correlation = df[columns].corr()

plt.figure(figsize=(10, 7))

sns.heatmap(
    correlation,
    annot=True,
    fmt=".2f",
    cmap="coolwarm"
)

plt.title("Correlation Between Laptop Features")
plt.show()
```

# Recommendations
### 1. Increase the Availability of Mid-Range Laptops

#### The majority of customers prefer laptops in the ₹40,000–₹80,000 price range. Brands should introduce more models in this segment to attract a larger customer base.

### 2. Focus on High-Performance Configurations

#### Laptops with Intel Core i5/Ryzen 5 processors, 16 GB RAM, and 512 GB SSD are the most popular. Manufacturers should prioritize these configurations to meet market demand.

### 3. Improve Value Instead of Increasing Discounts

#### If the analysis shows a weak relationship between discounts and ratings, companies should focus on improving product quality, battery life, and after-sales service rather than relying solely on discounts.

### 4. Expand the Premium Product Portfolio

#### Premium laptops generally receive higher customer ratings. Brands can introduce additional premium models with advanced features to target professionals and content creators.

### 5. Offer More Storage Options

#### Since 512 GB SSD is the most preferred storage capacity, brands should make it the standard configuration while also providing 1 TB SSD options for power users.

### 6. Improve Customer Satisfaction

#### Brands with comparatively lower ratings should analyze customer feedback and improve build quality, display, battery life, and customer support to enhance overall satisfaction.

### 7. Maintain Competitive Pricing

#### Brands should regularly compare their prices with competitors to remain competitive while ensuring customers receive good value for the specifications offered.

### 8. Promote High-Rated Products

#### Laptops with excellent customer ratings should receive greater visibility through featured listings, promotional campaigns, and recommendation sections to improve sales.
